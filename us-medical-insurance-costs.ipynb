```python
# create class Costumer

class Costumer:
    def __init__(self, identifier, age, sex, bmi, num_of_children, smoker):
        self.identifier = identifier
        self.age = age
        self.sex = sex
        self.bmi = bmi
        self.num_of_children = num_of_children
        self.smoker = smoker
        self.charged = 0
        self.insurance_estimated()
        self.insurance_diff_charged_estimated()
    
    def __repr__(self):
        return "The id of the current costumer is: {identifier}.".format(identifier=self.identifier)
    
    def insurance_charged(self, charged):
        self.charged = charged
            
    def insurance_estimated(self):
        self.estimated = 250 * self.age - 128 * self.sex + 370 * self.bmi + 425 * self.num_of_children + 24000 * self.smoker - 12500
        #print("ID {identifier}: The estimated insurance cost is {insurance_cost} dollars.".format(identifier = self.identifier, insurance_cost = self.estimated))
        return self.estimated
    
    def insurance_diff_charged_estimated(self):
        self.diff = round(self.charged - self.estimated, 1)
        return self.diff
        #if self.charged == self.estimated:
            #return "ID {identifier}: The charged insurance is equal to the estimated insurance cost.".format(identifier = self.identifier)
            
        #if self.charged > self.estimated:
            #return "ID {identifier}: The charged insurance is higher than the estimated insurance cost by: + {diff} dollars.".format(identifier = self.identifier, diff = self.diff)
        
        #if self.charged < self.estimated:
            #return "ID {identifier}: The charged insurance cost is lower than the estimated insurance cost by: {diff} dollars.".format(identifier = self.identifier, diff = self.diff)

    def has_insurance(self):
        if self.charged > 0:
            return "The costumer id {identifier} has insurance.".format(identifier=self.identifier)
        return "The costumer id {identifier} has no insurance yet.".format(identifier=self.identifier)
    
    def export_data(self):
        return "{identifier},{age},{sex},{bmi},{num_of_children},{smoker},{charged},{estimated_insurance},{diff}\n".format(
            identifier = self.identifier, age = self.age, sex = self.sex, bmi = self.bmi, 
            num_of_children = self.num_of_children, smoker = self.smoker, 
            charged = self.charged, estimated_insurance = self.estimated, diff = self.diff
        )
    

# create costumer 1
costumer1 = Costumer(0, 30, 1, 23.2, 0, 0)

# test methods of Costumer 
costumer1.insurance_charged(1200)

print(costumer1.charged)
print(costumer1.estimated)
print(costumer1.insurance_diff_charged_estimated())
print(costumer1.has_insurance())
```

```python
# convert the sex and smoker data to bool

def sex_data_conversion(sex):
    if sex == "female":
        sex_data_conversion = 1    
    else:
        sex_data_conversion = 0
    return sex_data_conversion

print(sex_data_conversion("female"))

def smoker_data_conversion(smoker):
    if smoker == "yes":
        smoker_data_conversion = 1
    else:
        smoker_data_conversion = 0
    return smoker_data_conversion

print(smoker_data_conversion("no"))
```

```python
import csv

# open csv data and save each costumer in costumer_data

costumer_data = []
with open("insurance.csv") as insurance_csv:
    reader = csv.DictReader(insurance_csv)
    identifier = 1
    for row in reader:              
        costumer = Costumer(identifier, int(row["age"]), sex_data_conversion(row["sex"]), round(float(row["bmi"]),1), 
                            int(row["children"]), smoker_data_conversion(row["smoker"]))
        costumer.insurance_charged(round(float(row["charges"]),1))
        costumer_data.append(costumer)
        
        identifier = identifier + 1
```

```python
# calculate the estimated insurance cost for each costumer on the csv data

def estimated_insurance_cost_by_costumer(costumer_list):
    for costumer in costumer_list:
        return costumer.insurance_estimated()
        #print(costumer.insurance_estimated())

def diff_insurance_by_costumer(costumer_list):
    for costumer in costumer_list:
        print(costumer.insurance_diff_charged_estimated())
```
